```python
import pandas as pd
```

```python
# Chargement du classement depuis le fichier CSV
df = pd.read_csv('input/france/ligue1_2024_2025.csv')
```

```python
# Création de features
df["goal_difference"] = df["G"] - df["GA"]
df["win_rate"] = df["W"] / df["M"]

# Génération des matchs (sans itertools)
matches = []

for i in range(len(df)):
    for j in range(len(df)):
        if i == j:
            continue  # on ne fait pas jouer une équipe contre elle-même
        
        home = df.iloc[i]
        away = df.iloc[j]
        
        home_score = home["goal_difference"] + home["win_rate"] * 10
        away_score = away["goal_difference"] + away["win_rate"] * 10

        if home_score > away_score + 2:
            prediction = "Home Win"
        elif away_score > home_score + 2:
            prediction = "Away Win"
        else:
            prediction = "Draw"

        matches.append({
            "home_team": home["TEAM"],
            "away_team": away["TEAM"],
            "home_score": round(home_score, 2),
            "away_score": round(away_score, 2),
            "prediction": prediction
        })

# Résultat
match_df = pd.DataFrame(matches)
match_df.head(10)
```

```python
# Définir les équipes du match que tu veux afficher
home_team = "Paris Saint Germain"
away_team = "Marseille"

# Filtrer le match
match_specifique = match_df[
    ((match_df["home_team"] == home_team) & (match_df["away_team"] == away_team)) |
    ((match_df["home_team"] == away_team) & (match_df["away_team"] == home_team))
]

# Afficher le match filtré
match_specifique
```
